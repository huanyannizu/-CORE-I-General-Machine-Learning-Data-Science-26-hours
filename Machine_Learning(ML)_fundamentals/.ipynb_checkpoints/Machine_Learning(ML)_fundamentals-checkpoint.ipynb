#### Exercise 1 - Basic Python programming

```python
#Importing Python modules
import numpy as np
import pandas as pd
from sklearn import datasets

# Load in the 'digits' data
digits = datasets.load_digits()
# Print the 'digits' data 

# Get the keys of the `digits` data
print(digits.keys())

# Print out the data
print(digits.data)

# Print out the target values
print(digits.target)
```

```python
# Isolate the `digits` data
digits_data = digits.data

# Inspect the shape
print(digits_data.shape)

# Isolate the target values with `target`
digits_target = digits.target

# Inspect the shape
print(digits_target.shape)

# Print the number of unique labels
number_digits = len(np.unique(digits.target))

# Isolate the `images`
digits_images = digits.images

# Inspect the shape
print(digits_images.shape)
```

```python
# Visualize the data images With matplotlib library

# Import matplotlib
import matplotlib.pyplot as plt

# Figure size (width, height) in inches
fig = plt.figure(figsize=(6, 6))

# Adjust the subplots 
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)

# For each of the 64 images
for i in range(64):
    # Initialize the subplots: add a subplot in the grid of 8 by 8, at the i+1-th position
    ax = fig.add_subplot(8, 8, i + 1, xticks=[], yticks=[])
    # Display an image at the i-th position
    ax.imshow(digits.images[i], cmap=plt.cm.binary, interpolation='nearest')
    # label the image with the target value
    ax.text(0, 7, str(digits.target[i]))

# Show the plot
plt.show()
```

```python
# visualize the target labels with an image

# Join the images and target labels in a list
images_and_labels = list(zip(digits.images, digits.target))

# for every element in the list
for index, (image, label) in enumerate(images_and_labels[:8]):
    # initialize a subplot of 2X4 at the i+1-th position
    plt.subplot(2, 4, index + 1)
    # Don't plot any axes
    plt.axis('off')
    # Display images in all subplots 
    plt.imshow(image, cmap=plt.cm.gray_r,interpolation='nearest')
    # Add a title to each subplot
    plt.title('Training: ' + str(label))

# Show the plot
plt.show()
```

#### Exercise 2 -  Simple statistics for a synthetic information data

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import datasets

# import iris data
iris = datasets.load_iris()
iris_data = pd.DataFrame(iris.data, columns=['sepal length','sepal width','petal length','petal width'])
y = iris.target
iris_data["target"] = [0 if label == 0 else 1 for label in y ] # combine class 1 and class 2

# split data into training set and testing set
X_train, X_test, Y_train, Y_test = train_test_split(iris_data[['sepal length','sepal width','petal length','petal width']], iris_data['target'], test_size=0.2, random_state=42)
print('number of training examples: ',X_train.shape[0])  
print('number of testing examples: ',X_test.shape[0]) 
print("") #start a new line
print(X_train.head()) #have a look at the first 5 rows of the training set
print("")
print(X_train.describe()) #compute descriptive statistics of the training
```

#### Exercise 3 -  Dimension reduction with PCA and t-SNE

3.1 PCA

```python
from sklearn import decomposition # import sklearn and matplotlib libraries

pca = decomposition.PCA(n_components=2) #create a PCA model with 2 components
principalComponents = pca.fit_transform(X_train) #fit PCA model. The resulting variable principalComponents contains the principal components.
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2']) #converting the resulting variable into a dataframe
print(principalDf.head())
```

```python
# visualize the training dataset after decomposition

import matplotlib.pyplot as plt # import matplotlib library for plotting
principalDf['Class'] = list(Y_train) # add label column to the principal dataframe for plotting purposes
targets = [0, 1]; colors = ['r', 'y']
for target, color in zip(targets,colors):
    indicesToKeep = principalDf['Class'] == target
    plt.scatter(principalDf.loc[indicesToKeep, 'principal component 1']
               , principalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , alpha=0.5)
plt.legend(targets)
plt.grid()
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('2 component PCA')
plt.show()
```

3.2 t-SNE

```python
import numpy as np
from sklearn.manifold import TSNE
X_embedded = TSNE(n_components=2).fit_transform(X_train)
X_embedded = pd.DataFrame(data = X_embedded, columns = ['embedding vector 1', 'embedding vector 2'])
print(X_embedded.shape)
print(X_embedded.head())
```

```python
# visualize the training dataset after decomposition

import matplotlib.pyplot as plt # import matplotlib library for plotting
X_embedded['Class'] = list(Y_train) 
targets = [0, 1]; colors = ['r', 'y']
for target, color in zip(targets,colors):
    indicesToKeep = X_embedded['Class'] == target
    plt.scatter(X_embedded.loc[indicesToKeep, 'embedding vector 1']
               , X_embedded.loc[indicesToKeep, 'embedding vector 2']
               , c = color
               , alpha=0.5)
plt.legend(targets)
plt.grid()
plt.xlabel('embedding vector 1')
plt.ylabel('embedding vector 2')
plt.title('decomposition with t-SNE')
plt.show()
```

### Exercise 4 - logistic regression to predict authentic (1)/inauthentic (0) on the banknote dataset

```python
# import LogisticRegression from sklearn library
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(solver='liblinear') #create logistic regression model with default parameter values
logreg.fit(X_train, Y_train) #fit the created model to training dataset
print('training accuracy:',logreg.score(X_train, Y_train)) #calculate training accuracy
print('testing accuracy:',logreg.score(X_test, Y_test))#calculate testing accuracy
```
